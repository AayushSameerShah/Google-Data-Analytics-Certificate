```python
trp_2019_Q1234 = trp_2019_Q1234.reindex(['trip_id', 'start_time', 'end_time', 'bikeid', 'tripduration', 
                         'from_station_id', 'from_station_name', 'to_station_id', 
                         'to_station_name', 'start_lat', 'start_lng', 'end_lat', 'end_lng',
                         'usertype', 'gender', 'birthday'], axis=1)
```

```python
# Dropping (not considering the birth year or gender as they are the Subscriber specific.)
trp_2019_Q1234.drop(["gender", "birthday"], axis= 1, inplace=True)
```

```python
trp_2019_Q1234 = trp_2019_Q1234.sort_values(by="start_time").reset_index(drop=True)
```
