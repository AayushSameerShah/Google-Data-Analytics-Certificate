# <center> 2020 </center>

```python
import os
```

```python
path = "Data/2020 — 2021/"
files = os.listdir(path)[1:]
```

```python
dfs = []
for file in files:
    dfs.append(pd.read_csv(path + file, low_memory=True))
```

Here we have the files spreaded in the monthly format. Each file for each month.

```python
trues = []
for i in range(len(dfs) - 1):
    trues.append((dfs[i].columns == dfs[i+1].columns).all())
```

```python
all(trues)
```

```python
trp_2020 = pd.concat(dfs).reset_index(drop=True)
```

```python
trp_2020
```

```python
trp_2020.started_at = pd.to_datetime(trp_2020.started_at, format="%Y-%m-%d %H:%M:%S")
trp_2020.ended_at = pd.to_datetime(trp_2020.ended_at, format="%Y-%m-%d %H:%M:%S")
```

```python
trp_2020.info()
```

```python
trp_2020_21 = trp_2020.sort_values(by="started_at")
```

```python
trp_2020_21.reset_index(drop=True, inplace=True)
```

```python
trp_2020_21.duplicated().sum()
```

```python
trp_2020_21[trp_2020_21.ride_id.duplicated()]
```

```python
trp_2020_21.to_csv("Data/CLEANED - YearWise/2020-21.csv", index=False)
```
